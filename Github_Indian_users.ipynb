```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution of joining years
plt.figure(figsize=(10, 6))
sns.histplot(df['joining_year'], bins=20, kde=True)
plt.title("Distribution of Joining Years")
plt.xlabel("Joining Year")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Plot distribution of repository counts
plt.figure(figsize=(10, 6))
sns.histplot(df['repositories'], bins=30, kde=True)
plt.title("Distribution of Repository Counts")
plt.xlabel("Number of Repositories")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Plot distribution of last year contributions
plt.figure(figsize=(10, 6))
sns.histplot(df['last_year_contributions'], bins=30, kde=True)
plt.title("Distribution of Last Year Contributions")
plt.xlabel("Contributions in Last Year")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Plot distribution of achievements
plt.figure(figsize=(10, 6))
sns.histplot(df['achievements_num'], bins=15, kde=True)
plt.title("Distribution of Achievements")
plt.xlabel("Number of Achievements")
plt.ylabel("Number of Users")
plt.show()

# Plot distribution of stars
plt.figure(figsize=(10, 6))
sns.histplot(df['stars'], bins=30, kde=True)
plt.title("Distribution of Stars")
plt.xlabel("Number of Stars")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Plot distribution of followers
plt.figure(figsize=(10, 6))
sns.histplot(df['followers'], bins=30, kde=True)
plt.title("Distribution of Followers")
plt.xlabel("Number of Followers")
plt.ylabel("Number of Users")
plt.show()

# Plot distribution of followings
plt.figure(figsize=(10, 6))
sns.histplot(df['followings'], bins=30, kde=True)
plt.title("Distribution of Followings")
plt.xlabel("Number of Followings")
plt.ylabel("Number of Users")
plt.show()
```

```python
# Compute correlation matrix
correlation_matrix = df.select_dtypes(include=np.number).corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of GitHub User Data")
plt.show()
```

```python
# Plotting contributions vs. followers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='last_year_contributions', y='followers', data=df, alpha=0.5)
plt.title("Contributions vs Followers")
plt.xlabel("Contributions in Last Year")
plt.ylabel("Number of Followers")
plt.show()
```

```python
# Plotting repositories vs stars
plt.figure(figsize=(10, 6))
sns.scatterplot(x='repositories', y='stars', data=df, alpha=0.5)
plt.title("Repositories vs Stars")
plt.xlabel("Number of Repositories")
plt.ylabel("Number of Stars")
plt.show()
```

```python
# Plotting followers vs followings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='followings', y='followers', data=df, alpha=0.5)
plt.title("Followings vs Followers")
plt.xlabel("Number of Followings")
plt.ylabel("Number of Followers")
plt.show()
```
